其中w是实值权重的向量,可以有很多变量对应的权重,w dot x 是点积运算,其中x也是一个向量,包含了许多需要加入的影响因素,设m是感知机的输入数量,b是偏置,偏置使得决策偏离原点,且不依赖任何输入值
这个方法的于神经网络的学习方式非常相似
- 初始化权重和偏差
- 前向传播
- 检查错误
- 反向传播并调整权重和偏置
- 所以训练重复这个过程

```python
def and_gate(x1, x2):
    w1, w2, th = 0.5, 0.5, 0.7
    result = x1 * w1 + x2 * w2
    if result > th:
        y = 1
    else:
        y = 0
    
    return y
```

```python
and_gate(0, 0)
```

```python
and_gate(0, 1)
```

```python
and_gate(1, 0)
```

```python
and_gate(1, 1)
```

```python
def and_gate2(x1, x2):
  w1, w2, th = 0.1, 0.1, 0.1
  result = x1 * w1 + x2 * w2
  y = 1 if result > th else 0
  return y
```

```python
and_gate2(1, 1)
```

```python
def and_gate3(x):
  w = np.array([0.5, 0.5])
  th = 0.7
  y = 1 if np.dot(x, w) > th else 0
  return y
```

```python
and_gate3(np.array([0.5, 0.5]))
```
